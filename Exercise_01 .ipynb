# Task 1

Calculate the sum of all natural numbers from $1$ to $n$, i.e. $1+2+\dots+n$, and return (not print) it.

## Solution

```python
def sum_to(n):    
    """A method that calculates the sum of all natural numbers from 1 to n.

    Args:
        n: Final integer

    Return:
        int: Sum of values 1 to n
    """
    # your code here
    sum=int(0.5*n*(n+1))
    return sum
    
```

# Task 2

Calculate the sum of $i$ to the power of $p$ of all natural numbers $i$ from $1$ to $n$, i.e. $1^p+2^p+\dots+n^p$, and return (not print) it. 

## Solution

```python
def sum_exp_to(n, p):    
    """A method that calculates the sum of i^p all natural numbers i from 1 to n.

    Args:
        n: Final integer
        p: Exponent

    Return:
        int: Sum of i^p for values i from 1 to n
    """
    # your code here
    Y=0
    for i in range(n+1):
        Y+= pow(i,p)

    return Y
```

## Example

```python
sum_exp_to(6, 2)
```

```python
assert sum_exp_to(4, 3) == 100, "your calculation is wrong"
```
