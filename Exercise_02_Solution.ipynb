# Task 1

The file **measurements.bin** is a binary file containing a list of measurements. Each line has exactly ten measures, while the first five measurements are in format signed integer, and the upcoming are unsigned integer values. An integer value requires 1 byte of memory. 

Extend the given `parse_measurements` method to parses the measurement file and return a tuple. The tuple's first entry should be the number of lines a given file contains, and the second value is the sum of a particular column. The column's index is specified by the "column" argument in the method. As already mentioned, the file contains exactly ten columns such that 0 <= column < 10. 

## Solution

```python
def parse_measurements(filename:str, column:int):    
    """A method that parses a binary file and returns the sum of a certain column

    Args:
        filename (str): Path as string to the corresponding binary file
        column (int): Index of the column that should be summed up

    Returns:
        tuple (int, int): Return a tuple with number of rows as first entry in 
                          tuple and the summed up value of a column as second value
    """
    column_sum = 0
    rows = 0
    with Path(filename).open('rb') as f:
        val = None
        while val != b'':               
            for i in range(10):    
                val = f.read(1)                
                if i == column:                    
                    column_sum += int.from_bytes(val, byteorder='big', signed=column < 5)                        
            rows += 1
    return rows - 1, column_sum
```

## Example

```python
parse_measurements('measurements.bin', 8)
```

# Task 2

In this task, you have to implement search functionality to search for a value by a specific key in a JSON file. As dataset you have to load the cars.json file, where each object consists of the following keys:

- Name
- Miles_per_Gallon
- Cylinders
- Displacement
- Horsepower
- Weight_in_lbs
- Acceleration
- Year
- Origin 

Extend the find_cars method to allow a user to search for a value in a given field. The return value of this function contains all car objects that match the search. Additionally, if a user tries to search for a key that is not part of the list above, the method should raise a KeyError exception.  Returning a  list of multiple cars is valid behavior. 

## Solution

```python
def find_cars(filename: str, search_field: str, search_term: any):
    """Implementation to search cars.json dataset by a
    given key for a given search term

    Args:
        filename (str): Path to cars.json file
        search_field (str): Field where to sarch for a term
        search_value (any): Search term

    Raises:
        KeyError: Occours if a search_field is not part of a record

    Returns:
        [dict]: A list of dictionaries that represents car objects
    """
    with Path(filename).open('r') as f:
        data = json.load(f)

    out = []    

    for e in data:
        if search_field not in e:
            raise KeyError(f"Invalid search_field of key \"{search_field}\"")

        if e[search_field] == search_term:
            out.append(e)

    return out
```

## Example

```python
result = find_cars('cars.json', "Cylinders", "8")
result
```
